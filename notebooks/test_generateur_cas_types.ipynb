```python
import matplotlib.pyplot as plt
from datetime import date
from openfisca_matplotlib.cas_types import create_scenario_inferieur_smic, create_scenario_superieur_smic
```

```python
import openfisca_france    # module décrivant le système socio-fiscal français
TaxBenefitSystem = openfisca_france.FranceTaxBenefitSystem  # Initialisation de la classe décrivant le système socio-fiscal français
tax_benefit_system = TaxBenefitSystem()  # Création d'une instance du système socio-fiscal français 
```

```python
scenario_kwargs = create_scenario_superieur_smic(year = 2014, nb_smic_max = 2, count = 20)
```

```python
scenario = tax_benefit_system.new_scenario().init_single_entity(**scenario_kwargs)
simulation = scenario.new_simulation()
```

```python
assert simulation.period.unit == 'year' 
assert simulation.period.start.year == 2014
assert simulation.period.start.month == 1
assert simulation.period.start.day == 1
assert simulation.period.size == 1
```

```python
from openfisca_matplotlib import graphs
```

```python
graphs.draw_bareme(simulation, x_axis = 'salaire_imposable')
```
